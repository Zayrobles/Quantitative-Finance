```python
#Import Data for statistical Analysis  

ticker = input("Enter stock ticker : ").upper()
period = "1y"

# fetch historical price data
historical_data = obb.equity.price.historical(symbol=f"{ticker}", period=period, provider='yfinance')
visible_data = historical_data.to_df() 

print((f"Historical Data for {ticker}, Time Range: {period}"))
print(f"Data Shape: {visible_data.shape}") 
print(visible_data.head())
```

```python
# Basic Statistical Summary

import pandas as pd
import numpy as np

# Basic Statistics
print(f"\n==== Basic Statistics for: {ticker}")
print(visible_data['close'].describe())

# Other Statistics 
print(f"\n Skewness: {visible_data['close'].skew():.4f}")
print(f"\n Kurtosis: {visible_data['close'].kurtosis():.4f}")
print(f"\n Standard Deviation: {visible_data['close'].std():.2f}")
print(f"\n Coefficient of Variation: {(visible_data['close'].std() / visible_data['close'].mean()) *100:.2f}")
```

```python
# Calculate Returns and Volatility

# Calculate daily returns 


visible_data['daily_return'] = visible_data['close'].pct_change()
visible_data['log_return'] = np.log(visible_data['close'] / visible_data['close'].shift(1))

print("\n** Returns  **")
print(f"Average Daily Return: {visible_data['daily_return'].mean():.4f} or {visible_data['daily_return'].mean()*100:.2f} %")
print(f"Daily Volatility: {visible_data['daily_return'].std():.4f} or {visible_data['daily_return'].std()*100:.2f}%")
print(f"Annualized Volatility: {visible_data['daily_return'].std()*np.sqrt(252)*100:.2f}%")
print(f"Annualized Return: {visible_data['daily_return'].mean()*252*100:.2f}%")

print("\n**Risk Metrics**")
print(f"Max Daily Gain:{visible_data['daily_return'].max()*100:.2f}%")
print(f"Max Daily Loss:{visible_data['daily_return'].min()*100:.2f}%")
print(f"95% (Value at Risk):{visible_data['daily_return'].quantile(0.05)*100:.2f}%")
print(f"99% (Value at Risk):{visible_data['daily_return'].quantile(0.01)*100:.2f}%")
```

```python
# Correlation Analysis (compared to the market S&P 500) 

spy_data = obb.equity.price.historical(symbol="SPY", period=period)
spy_df = spy_data.to_df()
spy_df['spy_return'] = spy_df['close'].pct_change()

#Merge the datasets

merged_df = pd.merge(visible_data[['daily_return']], spy_df[['spy_return']],
                     left_index=True, right_index=True, how='inner')

#Calculate Correlation

correlation = merged_df['daily_return'].corr(merged_df['spy_return'])
print("\n**Correlation with SPY**")
print(f"Correlation: {correlation:.4f}")
```
